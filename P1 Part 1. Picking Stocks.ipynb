```python
stock_prices = pd.read_csv('crsp_in_sample_2019spring.csv', parse_dates=['date'])
```

```python
stock_prices_sorted = stock_prices.sort_values(['cik','date'])
```

```python
display(stock_prices_sorted.head())
```

```python
stock_prices_pivoted = stock_prices_sorted.pivot(index="date",columns="ticker",values="adj_prc")
```

```python
display(stock_prices_pivoted.head())
display(stock_prices_pivoted.tail())
```

```python
stock_prices_pivoted.to_excel("Daily Stock Prices.xlsx")
```

```python
stock_prices_monthly = stock_prices_pivoted.resample(rule="M").mean()
```

```python
stock_prices_monthly.head()
```

```python
stock_prices_monthly.to_excel("Monthly Stock Prices.xlsx")
```

```python
risk_free_rate = 0.0021

tickers = list(stock_prices_monthly.columns.values)
sharpe_df = pd.DataFrame()

stocks_with_not_enough_data = [] 

for col in tickers:
    try: 
        number_of_months_with_data = stock_prices_monthly[col].dropna().shape[0]
        
        
        if(number_of_months_with_data==60):
        
            monthly_returns = stock_prices_monthly[col].pct_change()
            #print(monthly_returns)

            pct_changes = monthly_returns.values
            monthly_returns_vals = pct_changes[~np.isnan(pct_changes)]

            average_monthly_return = np.mean(monthly_returns_vals)
            #print(average_monthly_return)

            std_dev = np.std(monthly_returns_vals)
            #print(std_dev)

            sharpe_ratio = (average_monthly_return - risk_free_rate) / std_dev
            #print(sharpe_ratio)

            #sharpe_ratio[col] = sharpe_ratio

            sharpe_df = sharpe_df.append({"Ticker":col,"Average Monthly Returns":average_monthly_return,"Monthly Standard Deviation":std_dev,"Sharpe Ratio":sharpe_ratio}, ignore_index=True)
        else:
            stocks_with_not_enough_data.append(col)
    except Exception as e:
        print(e)
        sharpe_df = sharpe_df.append(None)
        
print("There were {} stocks without 60 months of full data".format(len(stocks_with_not_enough_data)))
```

```python
print("There are {} stocks total in the file".format(len(list(stock_prices_monthly.columns))))
```

```python
sharpe_df.sort_values("Sharpe Ratio",ascending=False, inplace=True)
```

```python
shares_outstanding_df = pd.read_excel("Shares_Outstanding_2014_12_31.xlsx")
shares_outstanding_df.shape
```

```python
shares_outstanding_df.head()
```

```python
shares_outstanding_df["Market Cap"] = shares_outstanding_df["Price or Bid/Ask Average"] * shares_outstanding_df["Shares Outstanding"] * 1000
```

```python
shares_outstanding_df.head()
```

```python
sharpe_and_marketcap_df = pd.merge(sharpe_df,shares_outstanding_df[["Ticker Symbol","Market Cap"]],left_on='Ticker',right_on="Ticker Symbol",how='left')
```

```python
stocks_with_highest_sharpe_and_over_three_billion_marketcap = sharpe_and_marketcap_df[sharpe_and_marketcap_df["Market Cap"] > 3000000000]
```
